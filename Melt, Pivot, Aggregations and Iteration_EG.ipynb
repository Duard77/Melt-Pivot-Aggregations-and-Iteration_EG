```python
""""
In this assignment you will experiment on your own. Using a health dataset of your choice (check with us if you are not sure), write code to demonstrate the following Pandas functions:

Melt
Pivot
Aggregation
Iteration
Groupby
"""
import pandas as pd
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/breast-cancer.data",
                names = ["Class", "Age", "Menopause", "Tumor size", "Inv-nodes", "Node-caps", "Deg-Malig", "Breast", "Breast quad", "Irridiat"])
```

```python
df.head()   #to see the structure of the df and the first 5 lines
```

```python
df.tail()   #to see the last 5 lines and see how many rows are in the data
```

```python
#Melt
Melt = pd.melt(df, id_vars = ["Class"], value_vars = ["Age"],
       var_name = "Demo", value_name = "D_Age range")   #Demo and D_Age range are the new column names created after the melt
Melt
```

```python
#Pivot
Pivot = df
Pivot['key'] = Pivot.groupby(["Breast", "Breast quad"]).cumcount()   #This line of code was needed to stop from getting "Index contains duplicate entries, cannot reshape"
Pivot = Pivot.pivot(index = ["key", "Breast"], columns = "Breast quad").reset_index()
Pivot
```

```python
#Aggregate
Aggregate = df.dropna(subset=['Age', 'Tumor size', 'Inv-nodes'])
Aggregate = Aggregate.aggregate({"Age":['min', 'max'],    #aggregate function mean will not work conventionally for range based data. Ranges would need to be spliced first
                          "Tumor size": ['min','max'],
                          "Inv-nodes": ['min', 'max']})
Aggregate
```

```python
#Iteration
Iteration = df
for index, row in Iteration.iterrows():   #This is helpful view of the record as a whole with the columns and data rearranged
    print(index, row)
```

```python
#Groupby
Groupby = df.groupby(['Breast', 'Breast quad']).count()  #This groups the data by Breast and Breast quadrant for the number of cases
Groupby
```
